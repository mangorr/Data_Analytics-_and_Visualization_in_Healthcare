# 5. Show the mean and standard deviation of the BMI of obese female patients (BMI 25.0 to <30).

```python
diabetes = (data[(data["BMI"]>=25.0) & (data["BMI"]<30) & (data["Female"]==True)])["BMI"]

if len(diabetes) > 0:
  diabetes_mean = diabetes.mean()
  diabetes_sd = diabetes.std()
    
print("Mean is : ", diabetes_mean)
print("Standard Deviation is : ", diabetes_sd)
```

# 7. Show the mean of the BMI of men who are obese (BMI over 30).

```python
obsedMen = (data[(data["BMI"]>30) & (data["Female"]==False)])["BMI"]

if len(obsedMen) > 0:
  obsedMen_mean = obsedMen.mean()
    
print("Mean is : ", obsedMen_mean)
```

# 9. Show the percentage of females with high blood pressure (SBP of 130 mm Hg or higher).

```python
highSBPFemale = len(data[(data["SBP"]>=130) & (data["Female"]==True)])
count = len(data[(data["Female"]==True)])
precentage = highSBPFemale / count

print("Percentage of females with high blood pressur", precentage)
```

# 10. Build a bar chart of the number of patients by the blood sugar levels (normal, prediabetes, and diabetes) at baseline.

## • A normal A1C level is below 5.7%.

## • A level of 5.7% to 6.4% indicates prediabetes.

## • A level of 6.5% or more indicates diabetes.

```python
normal = (data[(data["HemoglobinA1C.Baseline"])<5.7])["HemoglobinA1C.Followup"]
pre_diabetes = (data[(data["HemoglobinA1C.Followup"]>=5.7) & (data["HemoglobinA1C.Followup"]<=6.4)])["HemoglobinA1C.Followup"]
diabetes = (data[(data["HemoglobinA1C.Followup"])>=6.5])["HemoglobinA1C.Followup"]
```

```python
name_list = ['normal','pre_diabetes','diabetes']
num_list = [len(normal), len(pre_diabetes), len(diabetes)]

plt.bar(range(len(num_list)), num_list,tick_label=name_list)
plt.show()
```
